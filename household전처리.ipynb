```python
import numpy as np
# 카테고리에 대해서만 뽑은 데이터셋
household_category_data = household.iloc[:, 69:]

# 통계치 계산
descriptive_stats = household_category_data.describe()
min_values = descriptive_stats.loc['min'] #최소값
max_values = descriptive_stats.loc['max'] #최대값
median_values = descriptive_stats.loc['50%'] #중앙값

# 이상치 계산 (IQR 방식) : 일단 25% 아래 75% 위로 기준 잡았슘다~😯
Q1 = descriptive_stats.loc['25%']
Q3 = descriptive_stats.loc['75%']
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = {}
for column in household_category_data.columns:
    outlier_values = household_category_data[column][(household_category_data[column] < lower_bound[column]) | (household_category_data[column] > upper_bound[column])].values
    outliers[column] = outlier_values

# 결과 표로 출력
results = pd.DataFrame({
    '최소값': min_values,
    '최대값': max_values,
    '중앙값': median_values,
    '이상치': pd.Series(outliers)
})

print(results)
```

**[TODO 2] household 카테고리 스케일링**
 1. 각 카테고리별 스케일이 다 다르다. <br>
➡️ 데이터 정규화를 통해 스케일링을 하자. <br>
 https://soo-jjeong.tistory.com/122

원래는 `Min-Max 정규화` 방식을 사용하고자 했다. 이는 0과 1 사이의 범위로 변환해서 모든 카테고리가 동일한 스케일을 갖도록 하는 방법이다. <br>

> 정규화된 값 = (값-최소값) / (최대값 - 최소값)

이렇게 할 경우, 최소값에 0원이 너무 많아서 적합하지 않다. 다른 방법을 찾아봐야 했음!!
<br>
그 방법으로 2가지를 찾았다.

### 📍 **Z-스코어 정규화**
: 각 값에서 평균을 빼고, 그 결과를 표준편차로 나누어 계산하는 방식  <br>
✔️ 이 방식은 데이터 평균을 0, 표준편차를 1로 조정<br>
✔️ 이상치에 덜 민감하다는 특징 + 그래도 이상치에 영향을 받긴함

### 📍 **로버스트 스케일링**
: 중앙값과 IQR을 사용해서 데이터 스케일링
<br>
✔️ 이상치 영향을 최소화하는 데 효과적 <br>
✔️ 최소값, 최대값이 존재해도 중앙값 근처의 데이터 기준으로 스케일링되어서 안정적으로 분석 가능 <br>
✔️ 전반적인 분포 형태를 변경하지는 않고 스케일만 조정함. 패턴 보존 가능 <br>
✔️ 우리가 나중에 사용할 군집화, PCA, t-SNE 등 분석 사용 시 효과적 (∵ 기법들이 데이터 포인트 간의 거리에 기반을 두고 있음. 즉, 스케일의 통일성이 결과의 정확도도 높여줄 것으로 보임)

<br>
➡️ 로버스트 스케일링을 사용하는 게 이상치 영향을 받는 것을 최소화하면서 효과적일 것 같다. <br>
[사용해서 좋은 점]

*   이상치에 강하다(위 참조)
    *   극단적인 값들의 영향력은 감소할 수도 있다! 하지만 중요성을 소거한다는 의미가 아니다! 즉, 이상치들의 유의미함은 유지하면서도, 그게 전체 데이터 세트를 지배하는 것을 방지하도록 해주는 스케일링 기법이다.
*   예시: 의류 카테고리에서 대부분의 소비는 몇 만원대이지만 일부 고가의 아이템으로 인해 최댓값이 매우 높은 경우, 로버스트 스케일링을 통해서 극단적인 값을 조정하고, 대다수의 일반적인 소비 행태를 분석하는 데 좋다는 것임!!!

```python
from sklearn.preprocessing import RobustScaler
```

```python
#RobustScaler 객체 생성
scaler = RobustScaler()

#카테고리 선택
categories = ['의류', '미용서비스', '화장품','가전','인테리어','여행', '스포츠','문화생활']
householdCategory = household_drop[categories]

#스케일링 적용
robust_scaled_data = scaler.fit_transform(householdCategory)
```

```python
# 스케일링된 데이터 -> df 로 변환
robust_scaled_df = pd.DataFrame(robust_scaled_data, columns = [f"{cat}_scaled" for cat in categories])

#기존 household_drop에 추가
household_cleaned_todo2 = pd.concat([household_drop, robust_scaled_df], axis=1)
```
